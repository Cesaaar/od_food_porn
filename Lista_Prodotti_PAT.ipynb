# Lista Prodotti PAT - Prodotti Agroalimentari Tradizionali

Periodicamete il mipaaft pubblica l'aggiornamento dei prodotti PAT.. un misto tra pdf ed excel.
https://www.politicheagricole.it/flex/cm/pages/ServeBLOB.php/L/IT/IDPagina/398

Si definiscono "Prodotti Agroalimentari Tradizionali" (PAT) quei prodotti le cui metodiche di lavorazione, conservazione e stagionatura risultano consolidate nel tempo. In particolare, devono risultare praticate sul territorio di riferimento in maniera omogenea secondo regole tradizionali e protratte nel tempo, comunque per un periodo non inferiore ai 25 anni. Sono esclusi i prodotti agroalimentari registrati come Dop e Igp.

```python
import pandas as pd
import geopandas as gpd
import numpy as np
import folium #https://nbviewer.jupyter.org/github/python-visualization/folium/blob/master/examples
import json
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%pylab inline
```

```python
url = 'https://www.politicheagricole.it/flex/cm/pages/ServeAttachment.php/L/IT/D/9%252F1%252Fb%252FD.124faa78fe3caca619bb/P/BLOB%3AID%3D12327/E/xls'
df_pat = pd.read_excel(url, header=1)
df_pat.head(2)
```

```python
df_pat[['Regione/Prov. Autonoma','Totale']]
```

# Data Visualization
Utilizzo GeoPandas

```python
gdf = gpd.read_file("regioni.geojson")
gdf1=gdf.merge(df_pat[['Regione/Prov. Autonoma','Totale']], left_on='NOME_REG', right_on='Regione/Prov. Autonoma', how='left').fillna(0)
gdf1.head(2)
```

```python
gdf1
```

```python
from branca.colormap import linear

colormap = linear.YlGn_09.scale(
    gdf1['Totale'].min(),
    gdf1['Totale'].max())

colormap.caption = "Numero Prodotti PAT"
```

```python
# mappa
icon_url_abruzzo = 'http://www.ildatomancante.it/external/arrosticini_D7M_1.ico'
icon_url_piemonte = 'http://www.ildatomancante.it/external/prosciutto_val_vigezzo.ico'
icon_url_friuli = 'http://www.ildatomancante.it/external/friuli_fagiolo_laurons_Hau_icon.ico'
icon_url_lombardia = 'http://www.ildatomancante.it/external/lombardia_zucca_mantovana_9mu_icon.ico'
icon_url_valleaosta = 'http://www.ildatomancante.it/external/valle_d_aosta_micoula_qtL_icon.ico'
icon_url_veneto = 'http://www.ildatomancante.it/external/veneto_gallina_padovana_LqA_icon.ico'

centroid=gdf1.geometry.centroid
m=folium.Map(location=[centroid.y.mean(), centroid.x.mean()], zoom_start=5)

# layer 1
folium.GeoJson(gdf1[['geometry','NOME_REG','Totale']],
               name="Prodotti PAT in Italia",
               style_function=lambda x: {
                       "weight":2, 
                       'color':'black',
                       'fillColor':colormap(x['properties']['Totale']), 
                       'fillOpacity':0.5},
               highlight_function=lambda x: {'weight':3, 'color':'black'},
               smooth_factor=2.0,
               tooltip = folium.features.GeoJsonTooltip(
                                              fields=['NOME_REG','Totale',],
                                              aliases=['Regione','N Prodotti'], 
                                              labels=True, 
                                              sticky=True
                   )
              ).add_to(m)


#Abruzzo
popup_abruzzo = '<strong>Arrosticini</strong> ' + '<br>' + '<img src="'+icon_url_abruzzo+'" height="80" width="80">'
folium.Marker([42.3585, 13.9307], popup=popup_abruzzo).add_to(m)

#Piemonte
popup_piemonte = '<strong>Prosciutto Val Vigezzo</strong> ' + '<br>' + '<img src="'+icon_url_piemonte+'" height="80" width="80">'
folium.Marker([44.9959, 7.8662], popup=popup_piemonte).add_to(m)

#Valle d'Aosta
popup_valleaosta = '<strong>Micoula</strong> ' + '<br>' + '<img src="'+icon_url_valleaosta+'" height="80" width="80">'
folium.Marker([45.7522, 7.3828], popup=popup_valleaosta).add_to(m)

#Lombardia
popup_lombardia = '<strong>Zucca Mantovana</strong> ' + '<br>' + '<img src="'+icon_url_lombardia+'" height="80" width="80">'
folium.Marker([45.6601, 9.6899], popup=popup_lombardia).add_to(m)

#Veneto
popup_veneto = '<strong>Gallina Padovana</strong> ' + '<br>' + '<img src="'+icon_url_veneto+'" height="80" width="80">'
folium.Marker([45.6448, 11.9092], popup=popup_veneto).add_to(m)

#Friuli
popup_friuli = '<strong>Fagioli Laurons</strong> ' + '<br>' + '<img src="'+icon_url_friuli+'" height="80" width="80">'
folium.Marker([46.2255, 13.0957], popup=popup_friuli).add_to(m)

colormap.add_to(m)

m.save('FoodPorn_Prodotti_PAT.html')

m
```
